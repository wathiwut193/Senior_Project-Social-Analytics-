```python
from pythainlp.spell.pn import *
import time
```

```python
start = time.time()
read_text = open('print.txt','r',encoding='utf-8')
text = read_text.read()
x = text.split("|")

#print(x[0])
for i in x:
    print(correction(i))
read_text.close()

end = time.time()
print(end-start)
```

```python
start = time.time()
read_text = open('คลังข่าว/news_1_ตัด.txt','r+',encoding='utf-8')
text = read_text.read()
x = text.split("|")

#print(x[0])
for i in x:
    if i != " " and i != '' and i != "\n":
            print(correction(i))
        
read_text.close()

end = time.time()
print(end-start)
```
